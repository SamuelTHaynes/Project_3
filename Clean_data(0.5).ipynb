```python
import pandas as pd
```

```python
#import csv into dataframe
path = f'Project_3\\NRI_Table_Counties.csv'
raw_df = pd.read_csv(path)
raw_df.head()
```

```python
#select columns to keep
desired_columns = ['STATE', 'STATEABBRV', 'COUNTY', 'COUNTYTYPE', 'POPULATION', 'BUILDVALUE', 'AGRIVALUE', 'AREA', 'NRI_ID', 'TRND_EVNTS', 'TRND_EXPB', 'TRND_EXP_AREA', 'TRND_EALB', 'TRND_EALA', 'TRND_EALS', 'TRND_EALR', 'WFIR_EVNTS', 'WFIR_EXPB', 'WFIR_EXP_AREA', 'WFIR_EALB', 'WFIR_EALA', 'WFIR_EALS', 'WFIR_EALR', 'HRCN_EVNTS', 'HRCN_EXPB', 'HRCN_EXP_AREA', 'HRCN_EALB', 'HRCN_EALA', 'HRCN_EALS', 'HRCN_EALR']
only_cols = raw_df.loc[:, desired_columns]

#replace NaN with 'none'
only_cols['COUNTYTYPE'] = only_cols['COUNTYTYPE'].replace({pd.NA: 'none', 'NaN': 'none'})

#drop all non-county rows
countytype_df = only_cols[only_cols['COUNTYTYPE'].str.contains('County')]
countytype_df.reset_index(drop=True, inplace=True)
countytype_df.head()
```

```python
#drop County type col
clean_df = countytype_df.drop('COUNTYTYPE',axis=1)
clean_df.head()
```

```python
#check for Nan Values
Nan_values = clean_df.isna().sum()
Nan_values
```
